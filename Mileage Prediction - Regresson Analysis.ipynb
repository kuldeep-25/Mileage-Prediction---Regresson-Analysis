**Mileage** **Prediction** - **Regresson** **Analysis**

**Source**:

This dataset was taken from the StarLib library which is maintained at Carnegie Mellon University. The dataset was used in the 1983 American Statistical Association Exposition.

**Import** **Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Import** **Data**

```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv')
```

**Describe** **Data**

```python
df.head()
```

```python
df.nunique()
```

**Data** **Visualization**

```python
sns.pairplot(df, x_vars=['displacement', 'horsepower', 'weight', 'acceleration', 'mpg'], y_vars=['mpg']);
```

```python
sns.regplot(x = 'displacement', y = 'mpg', data = df);
```

**Data** **Preprocessing**

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df.info()
```

**Define Target Variable (y) and Feature Variables (X)**

```python
df.columns
```

```python
y = df['mpg']
```

```python
y.shape
```

```python
X = df[['displacement', 'horsepower', 'weight', 'acceleration']]
```

```python
X.shape
```

```python
X
```

**Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 2529 )
```

**Linear** **Regression** **Modeling**

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

**Mileage = 23.4 - 1.05Displacement-1.68Horsepower-4.10Weight-0.115Acceleration+error**

**Predict Test Data**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

**Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

**Polynomial Regression**

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
```

```python
X_train2 = poly.fit_transform(X_train)
```

```python
X_test2 = poly.fit_transform(X_test)
```

```python
lr.fit(X_train2, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
y_pred_poly = lr.predict(X_test2)
```

**Model Accuracy**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
mean_absolute_error(y_test, y_pred_poly)
```

```python
mean_absolute_percentage_error(y_test, y_pred_poly)
```

```python
r2_score(y_test, y_pred_poly)
```
